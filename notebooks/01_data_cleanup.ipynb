```python
import pandas as pd
import numpy as np
```

```python
df_cities = pd.read_csv('../data/city_attributes.csv')
df_cities
```

```python
df_humidity = pd.read_csv('../data/humidity.csv')
df_humidity = df_humidity.melt(
    id_vars=['datetime'],
    var_name='city',
    value_name='humidity'
)
df_humidity
```

```python
df_pressure = pd.read_csv('../data/pressure.csv')
df_pressure = df_pressure.melt(
    id_vars=['datetime'],
    var_name='city',
    value_name='pressure'
)
df_pressure
```

```python
df_temperature = pd.read_csv('../data/temperature.csv')
df_temperature = df_temperature.melt(
    id_vars=['datetime'],
    var_name='city',
    value_name='temperature'
)
df_temperature
```

```python
df_description = pd.read_csv('../data/weather_description.csv')
df_description = df_description.melt(
    id_vars=['datetime'],
    var_name='city',
    value_name='description'
)
df_description
```

```python
df_winddir = pd.read_csv('../data/wind_direction.csv')
df_winddir = df_winddir.melt(
    id_vars=['datetime'],
    var_name='city',
    value_name='wind_dir'
)
df_winddir
```

```python
df_windsp = pd.read_csv('../data/wind_speed.csv')
df_windsp = df_windsp.melt(
    id_vars=['datetime'],
    var_name='city',
    value_name='wind_sp'
)
df_windsp
```

```python
metrics = [df_description, df_humidity, df_pressure, df_temperature, df_winddir, df_windsp]

df = metrics[0]

for df_to_merge in metrics[1:]:
    df = df.merge(df_to_merge, on=['datetime', 'city'], how='outer')
```

```python
df
```

```python
df = df.merge(
    df_cities,
    left_on='city',
    right_on='City',
    how='outer'
)
df = df.drop('City', axis=1)
```

```python
df.to_csv('../data/combined.csv')
```
